```python
# XOR

import numpy as np

def sigmoid(net):
    return 1/(1+np.exp(-net))

def sigmoid_derivative(net):
    return sigmoid(net)*(1-sigmoid(net))

epochs = 130000
learning_rate = 0.1

inputs = np.array([[0,0],[0,1],[1,0],[1,1]])
outputs = np.array([[0],[1],[1],[0]])  # XOR için doğru çıkışlar

# Rastgele başlangıç ağırlıkları
input_size = 2
hidden_size = 2
output_size = 1

hidden_weights = np.random.uniform(size=(input_size, hidden_size))
output_weights = np.random.uniform(size=(hidden_size, output_size))
hidden_bias = np.random.uniform(size=(1, hidden_size))
output_bias = np.random.uniform(size=(1, output_size))

for epoch in range(epochs):
    total_error = 0

    for input, target in zip(inputs, outputs):
        # İleri Besleme
        hidden_net = np.dot(input, hidden_weights) + hidden_bias
        hidden_output = sigmoid(hidden_net)

        output_net = np.dot(hidden_output, output_weights) + output_bias
        final_output = sigmoid(output_net)

        # Hata Hesaplama
        error = target - final_output
        total_error += np.mean(error**2)

        # Geri Yayılım
        output_delta = learning_rate * error * sigmoid_derivative(output_net)
        hidden_error = np.dot(output_delta, output_weights.T)
        hidden_delta = learning_rate * hidden_error * sigmoid_derivative(hidden_net)

        # Ağırlık ve bias güncelleme
        output_weights += np.dot(hidden_output.T, output_delta)
        output_bias += output_delta
        hidden_weights += np.dot(input.reshape(-1,1), hidden_delta)
        hidden_bias += hidden_delta

    print(f'Epoch {epoch} - Error: {total_error}')

# Eğitilmiş modeli test et
print("Testing trained model on XOR inputs:")
for input in inputs:
    hidden_output = sigmoid(np.dot(input, hidden_weights) + hidden_bias)
    final_output = sigmoid(np.dot(hidden_output, output_weights) + output_bias)
    print(f'Input: {input}, Final Output: {final_output}')
```
